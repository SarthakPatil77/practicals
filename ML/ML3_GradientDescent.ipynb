```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return (x + 3) ** 2

def df(x):
    return 2 * (x + 3)

# Initial values
x_old = 2
precision = 0.0001
max_iters = 1000
learning_rate = 0.1

x_values = []
y_values = []


print("Iteration\tX Value\t\tf(X)\t\tGradient")
print("---------------------------------------------------")

for i in range(max_iters):
    grad = df(x_old)
    x_new = x_old - learning_rate * grad

    x_values.append(x_new)
    y_values.append(f(x_new))

    # Print each iterationâ€™s details
    print(f"{i+1}\t\t{x_new:6f}\t{f(x_new):.6f}\t{grad:.6f}")

    if abs(x_new - x_old) < precision:
        print("\nConverged at iteration:", i+1)
        break

    x_old = x_new
    
print(f"\nLocal minima occurs at x = {x_new:.4f}")
print(f"Minimum value of function y = {f(x_new):.4f}")

# Plot the descent path
plt.plot(x_values, y_values, 'o-', label='Gradient Descent Path')
plt.title("Gradient Descent to find local minima")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.legend()
plt.grid(True)
plt.show()
```
